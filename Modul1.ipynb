```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
"https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
"https://1.bp.blogspot.com/-03DDslWENIk/V7cOIpQyshI/AAAAAAAAAEc/RtD8Z-WuVu0tLT9bTzuM8uj_jWYicJIEgCLcB/s1600/anak-kucing-lucu-3.jpg"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (tinggi):
  image_3[int((tinggi)/2),y] =[255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis vertikal di tengah image
for x in range(tinggi):
    image_3[x, int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis diagonal dari kiri atas ke kanan bawah
for i in range(min(tinggi, lebar)):
    image_3[i, i] = [255,255,255]  # Garis dari kiri atas ke kanan bawah

# membuat garis diagonal dari kanan atas ke kiri bawah
for i in range(min(tinggi, lebar)):
    image_3[i, lebar - i - 1] = [255,255,255]  # Garis dari kanan atas ke kiri bawah

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

```

```python
# Konversi warna
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ukuran kotak putih
height, width = image.shape[:2]
box_size = 100  # Sesuaikan ukuran kotak sesuai kebutuhan

# Posisi kotak
top_left_x = (width - box_size) // 2
top_left_y = (height - box_size) // 2
bottom_right_x = top_left_x + box_size
bottom_right_y = top_left_y + box_size

# Menggambar kotak putih di tengah image_3
image_3[top_left_y:bottom_right_y, top_left_x:bottom_right_x] = [255, 255, 255]

# Menggabungkan gambar
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
